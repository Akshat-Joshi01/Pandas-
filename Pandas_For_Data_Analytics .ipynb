## 3 Data Structurer are used 
### Series: A one-dimensional array-like object with an index. Ex :- List As an argument . Syntax :- pd.Series(data,index)

### DataFrame: A two-dimensional labeled data structure with columns that can hold different data types. Ex :- List,Dictionary , Series ,Another DF .  As an argument. Syntax:-  pd.DataFrame(data) --> Most efficient to work with data sets 

### Panel: A three-dimensional(Multi Dimensional) labeled array-like container of data. (Less commonly used) .Ex :- (List,Dictionary , Series ,Another DF .  As an argument) Data, Major axis as (Columns),Minor axis (Rows). Syntax:-  pd.Panel(Data,Major axis,Minor axis)

# Mathamatical Operations

### - Addition (+): Adds values element-wise
### - Subtraction (-): Subtracts values element-wise 
### - Multiplication (*): Multiplies values element-wise
### - Division (/): Divides values element-wise
### - Exponentiation (**): Raises values to specified exponents element-wise
### - Absolute (abs()): Returns absolute value of each element
### - Square root (sqrt()): Calculates square root of each element
### - Logarithm (log()): Calculates natural logarithm of each element
### - Exponential (exp()): Calculates e raised to the power of each element
### - Maximum (max()): Returns maximum value across elements
### - Minimum (min()): Returns minimum value across elements
### - Sum (sum()): Calculates sum of all elements
### - Mean (mean()): Calculates mean of all elements
### - Median (median()): Calculates median of all elements
### - Standard Deviation (std()): Calculates standard deviation of elements
### Modulo (%) returns the remainder after division.
### Power (**) raises a number to a specified exponent.
### Less Than (<) checks if one value is less than another.
### Greater Than (>) checks if one value is greater than another.
### Equal To (==) checks if two values are equal.

# CONDITIONAL CHANGES

```python
d=pd.read_excel('C:\\Users\\Lenovo\\Desktop\\hidden folder\\DS , PY ,LIN\\Data analytics\\Pandas\\New Microsoft Excel Worksheet.xlsx')
df=pd.DataFrame(d)
df.drop_duplicates(inplace = True )
df
```

```python
df["Total_Marks"] = df["Telugu"] + df["English"] + df["Maths"] + df["Science"] + df["Social"]
df["Percentage"] = (df["Total_Marks"] / 500) * 100 # Assuming 100 marks per subject

print(df)
```

```python
df['Grade'] = 'None'
df
```

```python
'''
< 45 - FAIL 
> 45 & < 60 - PASS
>60 & < 80- FIRST CLASS
> 70 - DISTINCTION
'''
```

```python
df.loc[df['Percentage']< 45, ['Grade']] = 'FAIL'
df
```

```python
df.loc[(df['Percentage']>45) & (df['Percentage']<75),['Grade']]='PASS'
df
```

```python
df.loc[(df['Percentage']>60) & (df['Percentage']>80),['Grade']]='FIRST CLASS'
df
```
